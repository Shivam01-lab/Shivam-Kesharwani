```python
columns=["season","hr","holiday","weekday","weathersit","temp","hum","windspeed","casual","registered","cnt"]
data_merge[columns]=data_merge[columns].apply(pd.to_numeric)
```

```python
q1=data_merge["temp"].quantile(.25)
q3=data_merge["temp"].quantile(.75)
```

```python
def find_outliers(df,columns):
    q1= df[columns].quantile(.25)
    q3= df[columns].quantile(.75)
    iqr=q3-q1
    min_r=q1-1.5*iqr
    max_r=q3+1.5*iqr
    outliers_indices=df.index[(df[columns]<min_r) | (df[columns]>max_r)]
    return outliers_indices
```

```python
find_outliers(data_merge,"cnt")
```

```python
find_outliers(data_merge,"hr")
```

```python
find_outliers(data_merge,"temp")
```
